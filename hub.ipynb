```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import numpy as np
from PIL import Image
import requests
from io import BytesIO

# --- 1. C·∫§U H√åNH ---
# ƒê∆∞·ªùng d·∫´n t·ªõi model BEST c·ªßa b·∫°n (ki·ªÉm tra k·ªπ trong folder runs)
# V√≠ d·ª•: '/content/runs/classify/train/weights/best.pt'
MODEL_PATH = '/content/runs/classify/train2/weights/best.pt'

# Link ·∫£nh b·∫°n mu·ªën test (Ho·∫∑c thay b·∫±ng ƒë∆∞·ªùng d·∫´n file trong m√°y: 'test_image.jpg')
# V√≠ d·ª• l·∫•y m·ªôt ·∫£nh xe tƒÉng Leopard tr√™n m·∫°ng
IMAGE_URL = 'https://upload.wikimedia.org/wikipedia/commons/thumb/6/6f/Leopard_2_A7.JPG/1200px-Leopard_2_A7.JPG'

# --- 2. D·ª∞ ƒêO√ÅN ---
# Load model
model = YOLO(MODEL_PATH)

# T·∫£i ·∫£nh t·ª´ URL (n·∫øu d√πng ·∫£nh m·∫°ng)
response = requests.get(IMAGE_URL)
img = Image.open(BytesIO(response.content))

# Ch·∫°y d·ª± ƒëo√°n
results = model(img)

# --- 3. HI·ªÇN TH·ªä K·∫æT QU·∫¢ ---
# L·∫•y k·∫øt qu·∫£ ƒë·∫ßu ti√™n
result = results[0]
probs = result.probs  # X√°c su·∫•t c·ªßa c√°c l·ªõp
names = result.names  # T√™n c√°c l·ªõp

# L·∫•y class c√≥ x√°c su·∫•t cao nh·∫•t (Top-1)
top1_index = probs.top1
top1_conf = probs.top1conf.item()
class_name = names[top1_index]

print(f"\nüéØ K·∫øt qu·∫£ d·ª± ƒëo√°n: {class_name}")
print(f"üìä ƒê·ªô tin c·∫≠y: {top1_conf * 100:.2f}%")

# V·∫Ω ·∫£nh
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.title(f"PREDICTION: {class_name} ({top1_conf:.0%})", color='green', fontsize=15)
plt.axis('off')
plt.show()
```
